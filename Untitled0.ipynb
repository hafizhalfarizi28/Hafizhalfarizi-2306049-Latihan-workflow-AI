```python
import pandas as pd

data = {
    "Tanggal" : ["2025-02-05", "2025-02-05", "2025-02-05", "2025-02-05", "2025-02-05"],
    "Produk" : ["Bunga Lily", "Bunga Lavender", "Bunga Anggrek", "Bunga Mawar", "Bunga Matahari"],
    "Jumlah Terjual" : [20, 19, 18, 17, 16],
    "Stok Bunga" : [20, 21, 22, 23, 24],
    "Harga Perpaket" : [280000, 24000, 100000, 20000, 35000],
}

df = pd.DataFrame(data)
df.to_csv("Data_penjualan_bunga.csv", index=False)
print("File csv berhasil disimpan sebagai 'data_penjualan_bunga_latihan.csv'!")
```

```python
import pandas as pd

df = pd.read_csv("Data_penjualan_bunga.csv")
df["Tanggal"] = pd.to_datetime(df["Tanggal"])

df["Total Penjualan"] = df["Jumlah Terjual"] * df["Harga Perpaket"]
df["Keuntungan"] = df["Total Penjualan"] - (df["Jumlah Terjual"] * 10000)

print(df.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

df["Restock Bunga"] = df["Stok Bunga"] < 10
x = df[["Jumlah Terjual", "Stok Bunga"]]
y = df["Restock Bunga"] . astype(int)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = DecisionTreeClassifier()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Akurasi model:, {accuracy * 100:.2f}%")
```

```python
input_data = [[9,3]]
prediksi= model.predict(input_data)
status = "Restock Bunga" if prediksi[0] == 1 else "Restock Bunga"
print(f"Prediksi: {status}")
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["Jumlah Terjual"], df["Stok Bunga"], c=df["Keuntungan"], cmap="coolwarm", edgecolors="black")
plt.colorbar(label="Keuntungan")
plt.xlabel("Jumlah Terjual")
plt.ylabel("Stok Bunga")
plt.title("Hubungan antara jumlah terjual dan stok bunga bagi keuntungan")
plt.show()
```
